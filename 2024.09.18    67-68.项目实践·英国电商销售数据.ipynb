#### 一、项目介绍

**（1）整体流程:**

        下载数据--读取数据--评估数据--根据评估得到的结论解决发现的问题--每解决一个问题便再次查看清洗后的数据确保问题解决--保存清洗后的数据

**（2）数据集介绍:**

        这是一个跨国数据集，其中包含2010年12月12日至2011年12月9日期间发生的所有在英国注册的非商店在线零售业务的交易；
        该公司主要销售独特的全场礼品，并且大部分客户是批发商

**（3）字段含义:**

        数据形状为：542k 行x 8列，8个字段分别为发票号，发票日期，商品码，商品描述，数量，单价，顾客ID，国家
        
        
        InvoiceNo: 订单号码，每笔交易分配唯一的6位整数，而退货订单的代码以字母'c'开头
        
        StockCode: 产品代码，每个不同的产品分配唯一的5位整数
        
        Description: 产品描述，对每件产品的简略描述
        
        Quantity: 产品数量，每笔交易的每件产品的数量
        
        InvoiceDate: 订单日期和时间，每笔交易发生的日期和时间
        
        UnitPrice: 单价（英镑），单位产品价格
        
        CustomerID:顾客号码，每个客户分配唯一的5位整数
        
        Country: 国家的名字，每个客户所在国家/地区的名称

**（4）分析目标:**

        此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。
        本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 清洗数据

            根据前面评估部分得到的结论 我们需要进行的数据清理包括：

                        1.把InvoiceDate 变量的数据类型转换为日期时间
                        2.把CustomerID变量的数据类型转换为字符串
                        3.把description变量缺失的观察值删除
                        4.把country变量值"USA"替换为"United States"
                        5.把country变量值"UK"、"U.K."替换为”United Kingdom"
                        6.把Quantity变量值为负数的观察值删除
                        7.把Unitprice变量值为负数的观察值删除

            为了区分经过清理的数据和原始数据，我们创建新的变量cleand_data
            
            让它为original_data复制出的副本 我们之后的清理步骤将被运用在cleand_data上

```python
cleaned_data = original_data.copy()    # 复制副本
cleaned_data.head()                    # 提取副本的前五行
```

            1.把InvoiceDate 变量的数据类型转换为日期时间

```python
cleaned_data["InvoiceDate"] = pd.to_datetime(cleaned_data["InvoiceDate"])    # 调用.to_datetime()转换日期格式
cleaned_data.head()
```

            2.把CustomerID变量的数据类型转换为字符串

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].astype(str)    # 调用.astype(str) 转化为object字符串
cleaned_data["CustomerID"]
```

                    把CustomerID结尾的.0删除

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].str.slice(0, -2)    # .str.slice(保留的起点, 不保留的终点)
cleaned_data["CustomerID"]
```

            3.把Description变量缺失的观察值删除 并查看删除后该列空缺值个数和

```python
cleaned_data.dropna(subset=["Description"], inplace=True)    # .dropna(subset=["Description"])保留 "Description" 无缺失值的部分
cleaned_data["Description"].isnull().sum()                   # 求 "Description" 有缺失值的和----bool被视为1or0
```

            4.country变量值"USA"替换为"United States",并检查替换后"USA"变量值个数

```python
cleaned_data["Country"] = cleaned_data["Country"].replace({"USA": "United States"})    # 调用.replace({"老值 ": "新值 "}) 
len(cleaned_data[cleaned_data["Country"]== "USA"])
```

            5.把country变量值"UK"、"U.K."替换为"United Kingdom" ,并检查替换后"UK"，"U.K."数量

```python
cleaned_data["Country"] = cleaned_data["Country"].replace({"UK": "United Kingdom", "U.K.": "United Kingdom"} )

print(len(cleaned_data[cleaned_data["Country"]== "UK"]))
print(len(cleaned_data[cleaned_data["Country"]== "U.K"]))
```

            6.把Quantity变量值为负数的观察值删除并检查替换后Quantity变量值为负数的个数

```python
cleaned_data = cleaned_data[cleaned_data["Quantity"] >= 0]    # 反向操作：筛选出 "UnitPrice"] >= 0 的数据，并进行赋值
len(cleaned_data[cleaned_data["Quantity"] < 0])
```

            7.把UnitPrice变量值为负数的观察值删除并检查替换后UnitPrice变量值为负数的个数

            7.把UnitPrice变量值为负数的观察值删除并检查替换后UnitPrice变量值为负数的个数

```python
cleaned_data = cleaned_data[cleaned_data["UnitPrice"] >= 0]    # 同理进行反向操作：因为没学过删除特定行的命令
len(cleaned_data[cleaned_data["UnitPrice"] < 0])
```
